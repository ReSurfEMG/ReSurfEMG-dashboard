## Building test signal

We build 3 signals sapled at 50Hz, in a 1s time span: 
- signal_single_f has a single harmonic at 5Hz
- signal_double_f has two armonics at 5Hz and at 10Hz
- signal_continuous has two armonics at 5Hz and at 10Hz and a continuous componente at 0Hz

```python
baseline = 1
freq1 = 5
freq2 = 10
samp_rate = 50
time = np.linspace(0, 1, samp_rate)

signal_single_f = np.sin(2 * np.pi * freq1 * time)
signal_double_f = np.sin(2 * np.pi * freq1 * time)+np.sin(2 * np.pi * freq2 * time)
signal_continuous = np.sin(2 * np.pi * freq1 * time)+np.sin(2 * np.pi * freq2 * time)+baseline
```

```python
plt.plot(signal_single_f)
```

```python
plt.plot(signal_double_f)
```

```python
plt.plot(signal_continuous)
```

```python
y_c,x_c=hf.show_my_power_spectrum(signal_continuous, samp_rate, 25)
```

The power spectrum correctly shows the peaks at the frequencies used to create the signals. Nontheless there is a visualization artifact, due to the line connecting the last sample in the positive frequencies, to the first sample of the negative frequencies. The array of the frequencies is shown here below.

```python
plt.plot(x_c)
```

## Modifying visualization

A solution for the removal of the artifact, is to plot only the spectrum values in the positive frequencies domain, insted of limiting the x axis range.

```python
ind = [i for i,v in enumerate(x_c) if v >=0 ]
plt.plot(x_c[ind],np.abs(y_c[ind]))
```
